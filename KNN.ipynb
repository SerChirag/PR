```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as pl
from matplotlib import colors
from sklearn import neighbors, datasets
from collections import OrderedDict
cmaps = {}
cmaps

pl.rcParams['figure.figsize'] = [20, 16]
X = np.loadtxt("NLS_Group05/Class1.txt")
Y = [0]*len(X)
newX = np.loadtxt("NLS_Group05/Class2.txt")
Xnata1 = np.loadtxt("NLS_Group05/Class1.txt")
Xnata2 = np.loadtxt("NLS_Group05/Class2.txt")
newY = [1]*len(newX)
X = np.concatenate((X, newX))
Y = np.concatenate((Y, newY))
h = .02 # step size in the mesh


knn=neighbors.KNeighborsClassifier()

# we create an instance of Neighbours Classifier and fit the data.
knn.fit(X, Y)

# Plot the decision boundary. For that, we will asign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
x_min, x_max = X[:,0].min() - .5, X[:,0].max() + .5
y_min, y_max = X[:,1].min() - .5, X[:,1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
cmap = colors.ListedColormap(["#84FFFF", "#EEFF41"])
pl.pcolormesh(xx, yy, Z,cmap=cmap)

# Plot also the training points
pl.plot(Xnata1[:,0], Xnata1[:,1],"#2979FF",zorder=10,marker=".",linestyle="None")
pl.plot(Xnata2[:,0], Xnata2[:,1],"#388E3C",zorder=10,marker=".",linestyle="None")

pl.show()
```
