```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as pl
from matplotlib import colors
from sklearn import neighbors, datasets
from collections import OrderedDict

legend_facecolor='#f6ffc6'
legend_edgecolor='#ff07ff'
legend_linewidth=2
legend_loc='upper right'

c={}
c[1]='#54bdff' # blue Class Color
c[2]='#9ef442' # green
c[3]='#FFB74D' # red

c2={}
c2[1]='#2205c4' # Training Points Color
c2[2]='#388E3C'
c2[3]='#ad061f'
matplotlib.style.use('seaborn')

def knn():
    fig, ax1 = pl.subplots(1, 1, figsize=(14,8))
    
    pl.subplot(111)
    cmaps = {}
    cmaps
    pl.rcParams['figure.figsize'] = [20, 16]
    X = np.loadtxt("NLS_Group05/Class1.txt")
    Y = [0]*len(X)
    newX = np.loadtxt("NLS_Group05/Class2.txt")
    Xnata1 = np.loadtxt("NLS_Group05/Class1.txt")
    Xnata2 = np.loadtxt("NLS_Group05/Class2.txt")
    newY = [1]*len(newX)
    X = np.concatenate((X, newX))
    Y = np.concatenate((Y, newY))
    h = .02 # step size in the mesh

    
    knn=neighbors.KNeighborsClassifier()

    # we create an instance of Neighbours Classifier and fit the data.
    knn.fit(X, Y)

    # Plot the decision boundary. For that, we will asign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    x_min, x_max = X[:,0].min() - .5, X[:,0].max() + .5
    y_min, y_max = X[:,1].min() - .5, X[:,1].max() + .5
    X=np.linspace(-20,20,500)
    Y=np.linspace(-16,18,500)
    xx, yy = np.meshgrid(np.arange(-20, 20, h), np.arange(-16,18,h))
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    cmap = colors.ListedColormap([c[1], c[2]])
    pl.pcolormesh(xx, yy, Z,cmap=cmap)

    # Plot also the training points
    pl.plot(Xnata1[:,0], Xnata1[:,1],c2[1],zorder=10,marker=".",linestyle="None")
    pl.plot(Xnata2[:,0], Xnata2[:,1],c2[2],zorder=10,marker=".",linestyle="None")

    pl.title('Class1 vs Class2', fontsize=16)
    
    colo = [c[1],c2[1],c[2],c2[2]]
    texts = ["Class 1","Training 1", "Class 2", "Training 2"]
    
    handles = [pl.Rectangle((0,0),1,1,color=colo[i]) for i  in range(len(texts))]
    legend=fig.legend(handles=handles, labels=texts,loc='right', ncol=1,frameon=1,
                      labelspacing=1.5, borderpad=0.75, prop=dict(weight='black',size='large'))
    
    frame = legend.get_frame()
    frame.set_facecolor(legend_facecolor)
    frame.set_edgecolor(legend_edgecolor)
    frame.set_linewidth(legend_linewidth)
#     pl.show()
```

```python
knn()
```
